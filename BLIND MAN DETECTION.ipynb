```python
total_normal=len(os.listdir("Video\\Normal\\"))
total_blind=len(os.listdir("Video\\Blind\\"))
total_normal
total_blind
```

```python
class_name = "Normal"
for i in range(total_normal+total_blind):
    if(i<int(total_normal)):
        class_name = "Normal"
        source="Video\\Normal\\" + str(i+1) +".mp4"
    else:
        class_name = "Blind"
        source="Video\\Blind\\" + str(int(i+1-(total_normal))) +".mp4"



    print(i+1)
    cap = cv2.VideoCapture(source)
    
    # Initiate holistic model
    with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    
        while cap.isOpened():
            ret, frame = cap.read()
            if ret==0:
                break
            frame=cv2.resize(frame, [640,480])
            # Recolor Feed
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False        
        
            # Make Detections
            results = holistic.process(image)
            # print(results.face_landmarks)
        
           
        
            # Recolor image back to BGR for rendering
            image.flags.writeable = True   
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        

            # Export coordinates
            try:
                # Extract Pose landmarks
                pose = results.pose_landmarks.landmark
                pose_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in pose]).flatten())
        
                # Concate rows
                row = pose_row  
                
                # Append class name 
                row.insert(0, class_name)
            
                # Export to CSV
                with open('TRY1.csv', mode='a', newline='') as f:
                    csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                    csv_writer.writerow(row) 
            
            except:
                pass
                                
         
    cap.release()
    cv2.destroyAllWindows()

    
    
```

# 3. Train Model Using Scikit Learn

## 3.1 Read in Collected Data and Process

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('DATA.csv')
```

```python
X = df.drop('class', axis=1) # features
y = df['class'] # target value
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

```python
len(y_train)
```

## 3.2 Train Machine Learning Classification Models

```python
from sklearn.pipeline import make_pipeline 
from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA

from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
```

```python
pipelines = {
    'lr':make_pipeline(PCA(0.99),StandardScaler(),LogisticRegression(max_iter=1500)),
   # 'sv1':make_pipeline(StandardScaler(), SVC(gamma='auto',max_iter=-1)),
   # 'sv2':make_pipeline(PCA(0.99),StandardScaler(), SVC(gamma='auto',max_iter=-1)),
   # 'ls1':make_pipeline(StandardScaler(),LinearSVC(random_state=0, tol=1e-5,max_iter=15000)),
   # 'ls2':make_pipeline(PCA(0.99),StandardScaler(),LinearSVC(random_state=0, tol=1e-5,max_iter=15000)),
   # 'sg1':make_pipeline(StandardScaler(),SGDClassifier(max_iter=1500, tol=1e-3,)),
   # 'sg2':make_pipeline(PCA(0.99),StandardScaler(),SGDClassifier(max_iter=1500, tol=1e-3,)),
    'rc':make_pipeline(PCA(0.99),StandardScaler(), RidgeClassifier(max_iter=1500)),
    'rf':make_pipeline(PCA(0.99),StandardScaler(), RandomForestClassifier()),
    'gb':make_pipeline(PCA(0.99),StandardScaler(), GradientBoostingClassifier()),
    'nn':make_pipeline(PCA(0.99),StandardScaler(), MLPClassifier(hidden_layer_sizes=(4,4), activation='relu', solver='adam', max_iter=1500)),
    }
```

```python
fit_models = {}  #TRANING MODEL
i=1
for algo, pipeline in pipelines.items():
    print(i)
    model = pipeline.fit(X_train, y_train)
    fit_models[algo] = model
    i=i+1
```

```python
fit_models['nn']
```

```python
fit_models
```

```python
fit_models['lr'].predict(X_test)
```

## 3.3 Evaluate and Serialize Model 

```python
from sklearn.metrics import accuracy_score  
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
import pickle
```

```python
for algo, model in fit_models.items():
    yhat = model.predict(X_test)
    print("\nAlgorithm:            ",algo)
    print("Accuracy Score          ",algo, accuracy_score(y_test, yhat))
    print("F1 Score                ",algo,f1_score(y_test,yhat, average='micro'))
    print("Precision Score         ",algo,precision_score(y_test,yhat, average='micro'))
    print("Recall Score            ",algo,recall_score(y_test,yhat, average='micro'))
    print("\nClassification Report   ")
    print(classification_report(y_test,yhat))
```
